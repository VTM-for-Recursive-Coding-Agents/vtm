# Running Benchmarks
Use the orchestrator script with the supported flags. For SWE-bench in `eval-only` mode, set a valid predictions file path first.

Example below runs both benchmarks with baseline provider, compares by model, and uses one shared run id prefix.

```python
PROJECT_ROOT="$(cd "$(dirname "${BASH_SOURCE[0]}")/.." && pwd)"

LCB_MODEL="gpt-4-1106-preview"
SWE_PRED_PATH="$PROJECT_ROOT/benchmarks/SWE-bench/outputs/your_predictions.jsonl"

cd "$PROJECT_ROOT/scripts" && ./run_benchmarks.sh \
  --benchmark all \
  --provider baseline \
  --lcb-model "$LCB_MODEL" \
  --swe-mode eval-only \
  --swe-predictions-path "$SWE_PRED_PATH"
```
